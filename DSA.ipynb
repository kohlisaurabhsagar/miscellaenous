```python
# Binary Search
# mainly used when a sorted array is given
# ---------------------------------------------------------------------------------------------------------------
def binSearch(arr, target):
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = start + (end-start)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return -1
# ---------------------------------------------------------------------------------------------------------------
# FIRST AND LAST OCCURENCE OF A NO.

def firstOccur(arr, target) -> int:
    result = -1
    start = 0
    end = len(arr) - 1

    while start <= end:
        mid = start + ((end - start) // 2)

        if target == arr[mid]:
            result = mid
            end = mid - 1  
        elif target < arr[mid]:
            end = mid - 1
        else:
            start = mid + 1

    return result

def lastOccur(arr, target) -> int:
    result = -1
    start = 0
    end = len(arr) - 1

    while start <= end:
        mid = start + ((end - start) // 2)

        if target == arr[mid]:
            result = mid
            start = mid + 1  
        elif target < arr[mid]:
            end = mid - 1
        else:
            start = mid + 1

    return result


def countOccur(arr,target) -> int:
    start = firstOccur(arr, target)
    end = lastOccur(arr, target)
    result = end - start + 1
    return result

# ROTATED ARRAY
# no of rotations = minimum/maximum no.'s index depends on if array is orted ascedning or descending
def numRotation(arr) -> int:
    start = 0
    end = len(arr) - 1

    while start <= end:
        mid = start + (end - start)//2
        prev = (mid + len(arr) - 1) % len(arr)
        next = (mid + 1) % len(arr)
        
        if arr[mid] <= arr[end] and arr[start] <= arr[mid]:
            return start      
        
        if arr[mid] < arr[prev] and arr[mid] < arr[next]:  # for array sorted in ascending order
            return mid # return len(arr) - mid if array is left rotated
        # if arr[mid] > arr[prev] and arr[mid] > arr[next]:  # for array sorted in descending order
        #     return  mid  
        if arr[start] <= arr[mid]: 
            start = mid + 1      # end = mid - 1 
        elif arr[end] >= arr[mid]:
            end = mid - 1   # start = mid + 1

# FINDING ELEMENT IN ROTATED ARRAY
def findNum(arr, target) -> int:
    start = 0
    end = len(arr) - 1
    temp = -1

    while start <= end:
        mid = start + (end - start)//2
        prev = (mid + len(arr) - 1) % len(arr)
        next = (mid + 1) % len(arr)      
        
        if arr[mid] < arr[prev] and arr[mid]:  # for array sorted in ascending order
            temp = mid
        if arr[start] <= arr[mid]: 
            start = mid + 1      # end = mid - 1 
        elif arr[end] >= arr[mid]:
            end = mid - 1   # start = mid + 1

    result1 = binSearch(arr[:temp], target)
    result2 = binSearch(arr[temp:], target)
    if result1 > result2:
        return result1
    else:
        return result2 + len(arr[:temp]) 


# FINDING ELEMENT IN NEARLY SORTED ARRAY
def modbinSearch(arr, target):
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = start + (end - start)//2
        if arr[mid] == target:
            return mid
        elif arr[mid - 1] == target and (mid-1) >= start:
            return mid - 1
        elif arr[mid + 1] == target and (mid+1) <=end:
            return mid + 1
        elif target > arr[mid]:
            start = mid + 1
        else:
             end = mid - 1
    return -1

# FINDING THE FLOOR or CEIL VALUE OF AN ELEMENT IN ARRAY
# if the element is not present then finding the highest no. less than it present in array for floor element nad finding the lowest no. hiher than the target  present in array.
def floorElement(arr, target) -> int:
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = start + (end - start)//2
        if arr[mid] > target:
            end = mid - 1
        elif arr[mid] < target:
            start = mid + 1  
        else:
            return arr[mid]       
    return arr[end] if end < len(arr) and arr[end] < target else -1

def ceilElement(arr, target) -> int:
    start = 0
    end = len(arr) - 1
    while start <= end:
        mid = start + (end - start)//2
        if arr[mid] > target:
            end = mid - 1
        elif arr[mid] < target:
            start = mid + 1  
        else:
            return arr[mid]     
    return arr[start] if start < len(arr) and arr[start] > target else -1
            



# ---------------------------------------------------------------------------------------------------           

if __name__ == "__main__":
    # -----------------------------------------------------------
    # inpArr = [1, 5, 7, 9, 9, 9, 13, 16, 17]
    # print(firstOccur(inpArr, 9))
    # print(lastOccur(inpArr, 9))
    # print(countOccur(inpArr, 9))
    # -----------------------------------------------------------
    # inpArr = [19, 13, 10, 9, 7, 31, 29, 28, 27]
    # inpArr = [6, 7, 8, 1, 2, 3, 4, 5]
    # print(numRotation(inpArr))
    # -----------------------------------------------------------
    # inpArr = [7, 8, 9, 1, 2, 3, 4, 5, 6]
    # print(findNum(inpArr, 7))
    # -----------------------------------------------------------
    # inpArr = [0, 1, 2, 3, 4, 5, 8, 6, 7, 9]
    # print(modbinSearch(inpArr,8))
    # -----------------------------------------------------------
    # inpArr = [1, 4, 5, 9]
    # print(ceilElement(inpArr, 8))
    # print(floorElement(inpArr, 2))
    # -----------------------------------------------------------
```

```python
#PRIORITY QUEUE USING LIST

class priorityQueue():
  def __init__(self):
        self.pQueue = []


  def enqueue(self, element):
        self.pQueue.append(element)
        print("Appended element:", element)
        print(self.pQueue)

  def dequeue(self):
        if self.isEmpty():
            print("Empty pQueue")
        else:
            self.pQueue.sort()  #giving priority to lower value
            e = self.pQueue.pop(0)
            print("Removed element:", e)
            print(self.pQueue)

  def isEmpty(self):
        if len(self.pQueue) == 0:
            return True
        else:
            return False

A = priorityQueue()
A.enqueue(15)
A.enqueue(7)
A.enqueue(29)
A.enqueue(8)
A.dequeue()
A.isEmpty()
```

```python
#linkedList 
class linkedList():
    def __init__(self) -> None:
        self.head = None
    
    def print_ll(self):
        if self.head is None:
            print("Empty LinkedList!!!")
        else:
            n = self.head 
            while n is not None:
               print(n.data,"->",end=" ")
               n = n.ref 
    def add_begin(self, data):
        new_node = Node(data)
        new_node.ref = self.head
        self.head = new_node
        
    def add_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head                  
            while n.ref is not None:
                n = n.ref
            n.ref = new_node
    
    def add_inbetween(self, data, index):   #can be done in this way too.
        new_node = Node(data)
        i = 1
        n = self.head
        while i < index:
            n = n.ref
            i+=1
        new_node.ref = n.ref
        n.ref = new_node

    def after_node(self, data, node_value):
        new_node = Node(data)
        n = self.head
        while n is not None:
          if n.data == node_value :
            new_node.ref = n.ref
            n.ref = new_node
            break
          n = n.ref
        if n is None:
            print("Node not Present!!!")

    def before_node(self,data, node_value):
        new_node = Node(data)
        if self.head is None:     
            print("Empty Linked List!!!")
            return
        if self.head.data == node_value:
            new_node.ref = self.head
            self.head = new_node
            return
        else: 
            n = self.head
            while n.ref is not None:
                if n.ref.data == node_value:    #to know the data of the next node
                    break
                n = n.ref
            if n.ref is None:
                print("Given Node Value not Found.")
            else:
                new_node.ref = n.ref
                n.ref = new_node  
                return
    def insert_empty(self,data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("Non-Empty LinkedList!!!")

    def del_begin(self):
        if self.head is None:
            print("LinkedList already Empty!!!")
        else:
            self.head = self.head.ref
    
    def del_last(self):
        if self.head is None:
            print("LinkedList already Empty!!!")
        elif self.head.ref is None: #if there's only one element in linked list
            self.head = None
        else:
            n = self.head
            while n.ref.ref is not None:  #to know th reference of the next node
                n = n.ref
            n.ref = None
        
    def del_node(self,node_value):
        if self.head is None:
            print("LinkedList already Empty!!!")
        else:
            n = self.head
            if self.head.data == node_value:
                self.head = self.head.ref
            #while n.ref.data != node_value:   # to know the data of the next node
                #n = n.ref
                #if n.ref.ref is None:         # if element not found
                    #print("Element Not Found!!!") 
                    #break           
            #if n.ref.ref is None:
                #n.ref = None
            #else:
                #n.ref = n.ref.ref       #deleting the middle node     
            while n.ref is not None:
                if n.ref.data == node_value:
                    break
                n = n.ref
            if n.ref is None:
                print("Element Not Found!!!") 
            else:
                n.ref = n.ref.ref  
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None

    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent

        return level

    def printTree(self):
        spaces = ' ' * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.data)
        if self.children:
            for child in self.children:
                child.printTree()

    def add_child(self, child):
        child.parent = self
        self.children.append(child)


def build_product_tree():
    root = TreeNode("Electronics")

    laptop = TreeNode("Laptop")
    laptop.add_child(TreeNode("Mac"))
    laptop.add_child(TreeNode("Surface"))
    laptop.add_child(TreeNode("Thinkpad"))

    cellphone = TreeNode("Cell Phone")
    cellphone.add_child(TreeNode("iPhone"))
    cellphone.add_child(TreeNode("Google Pixel"))
    cellphone.add_child(TreeNode("Vivo"))

    tv = TreeNode("TV")
    tv.add_child(TreeNode("Samsung"))
    tv.add_child(TreeNode("LG"))

    root.add_child(laptop)
    root.add_child(cellphone)
    root.add_child(tv)

    return root

if __name__ == '__main__':
    root = build_product_tree()
    root.printTree()
```

```python
class BST:
    def __init__(self, key) -> None:
        self.key = key
        self.lchild = None
        self.rchild = None
    
    def insert(self, data):
        if self.key is None:
            self.key = data
            return
        if self.key == data:
            return
        if self.key > data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = BST(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = BST(data)
        
    def search(self, data):
        if self.key == data:
            print("Node Found!!!")
            return
        if self.key > data:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Node Not Found!!!")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Node Not Found!!!")
        
    def preorder(self):
        print(self.key, end = " ")
        if self.lchild:
            self.lchild.preorder()
        if self.rchild:
            self.rchild.preorder()

    def inorder(self):
        if self.lchild:
            self.lchild.inorder()
        print(self.key, end = " ")
        if self.rchild:
            self.rchild.inorder()
    
    def postorder(self):
        if self.lchild:
            self.lchild.postorder()
        if self.rchild:
            self.rchild.postorder()
        print(self.key, end=" ")

    # def count(self,node):
        # if node is None:
            # return 0
        # return 1 + self.count(node.lchild) + self.count(node.rchild)
    
    def delete(self,data):
        if self.key is None:
            print("Tree is EMPTY!!!")
            return
        
        if data < self.key:
            if self.lchild:
                self.lchild = self.lchild.delete(data)
            else:
                print("Node Not Present in Tree!")
                return self
        elif data > self.key:
            if self.rchild:
                self.rchild = self.rchild.delete(data)
            else:
                print("Node Not Present in Tree!")
                return self  
        else:
            if self.lchild is None:
                temp = self.rchild
                if self.key == data:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return
                self = None
                return temp
            if self.rchild is None:
                temp = self.lchild
                if self.key == data:
                    self.key = temp.key
                    self.lchild = temp.lchild
                    self.rchild = temp.rchild
                    temp = None
                    return
                self = None
                return temp
            # when there are two children either the parent node(which is gonna be deleted) can be replaced with highest element of lchild or smallest of rchild.
            # here smallest of rchild is replaced
            node = self.rchild
            while node.lchild:
                node = node.lchild
            self.key = node.key
            self.rchild = self.rchild.delete(node.key)    
            # here largest of lchild is replaced
            # node = self.lchild
            # while node.rchild:
                # node = node.rchild
            # self.key = node.key
            # self.lchild = self.lchild.delete(node.key)
        return self
    
    def max_node(self):
        if self.key is None:
            print("Tree is EMPTY!!!")
            return
        if self.rchild is None:
            print(self.key)
        else:
            current_node = self.rchild
            while current_node.rchild:
                current_node = current_node.rchild
            print(current_node.key)

    def min_node(self):
        if self.key is None:
            print("Tree is EMPTY!!!")
            return
        if self.lchild is None:
            print(self.key)
        else:
            current_node = self.lchild
            while current_node.lchild:
                current_node = current_node.lchild
            print(current_node.key)
            
        
        

          
ls = [45,55,54,1,87,96,56,74]
Btree = BST(9)
for i in ls:
    Btree.insert(i)
Btree.search(1) 
Btree.preorder()
print(" -> Preorder")   
Btree.inorder()
print(" -> Inorder")   
Btree.postorder()
print(" -> Postorder")
Btree.delete(108)
Btree.inorder()
print("\n-------------------")
# Btree.delete(87)   causing error
# Btree.inorder()
# print("\n-------------------")
Btree.max_node()
Btree.min_node()
```

```python
#Knapsack Problem -> Maximize the knapsack value.
weight = [24, 58, 23, 78, 12, 45, 23, 95, 45, 69, 82, 75]
value = [200, 400, 564, 759, 354, 426, 354, 154, 964, 634, 456, 851]
capacity = 108
item_count = 12


# -----------------------------------------------------------------------------------------------------------------------------------------------
# NAIVE APPROACH (AS WE ARE CALCULATING ALL THE POSSIBLE COMBINATIONS) {CAN BE IMPLEMENTED USING RECURSIVE OR ITERATIVE APPPROACH}
# -----------------------------------------------------------------------------------------------------------------------------------------------

def knapsackN(weight, value, capacity, item_count):
    if item_count == 0 or capacity == 0:
        return 0
    
    # if (weight[item_count-1]<capacity):
        # return max(
            # value[item_count-1] + knapsackN(weight[:item_count-1],value[:item_count-1],capacity-weight[item_count-1],item_count-1),
            # knapsackN(weight[:item_count-1],value[:item_count-1],capacity,item_count-1)
        # )

    if weight[item_count - 1] <= capacity:
        return max(
            value[item_count - 1] + knapsackN(weight, value, capacity - weight[item_count - 1], item_count - 1),
            knapsackN(weight, value, capacity, item_count - 1)
        )
    
        #when weight[item_count-1]>capacity
    else:
        # return knapsackN(weight[:item_count-1],value[:item_count-1],capacity,item_count-1)
        return knapsackN(weight, value, capacity, item_count - 1)


# -------------------------------------------------------------------------------------------------------------------------------------------------
# MEMORIZATION (TOP DOWN APPROACH) {RECURSIVE} WE WILL STORE ALL THE VARIABLE VALUES(i.e. KNAPSACK WEIGHT & NO. OF ITEMS) IN 2-D ARRAYS
# The top-down solution, commonly known as the memoization technique, is an enhancement of the recursive solution. It overcomes the problem 
# of calculating redundant solutions over and over again by storing them in a table.
# In the recursive approach, the two variables that kept changing in each call were the total weight of the knapsackN and the number of ITEMS we had.
# So, we’ll use these two variables to define each distinct subproblem. Therefore, we need a 2-D array to store these values and the result of any 
# given subproblem when we encounter it for the first time. At any later time, if we encounter the same sub-problem, we can return the stored result 
# from the table with an O(1)O(1) lookup instead of recalculating that subproblem.
# --------------------------------------------------------------------------------------------------------------------------------------------------

# def find_knapsack(capacity, weights, values, item_count):
dp = [[0 if i==0 or j==0 else -1 for j in range(capacity + 1)] for i in range(item_count + 1)]
    # return find_knapsack_value(capacity, weights, values, item_count, dp)

def knapsackM(weights, values, capacity, item_count, dp):
    # Base case
    if item_count == 0 or capacity == 0:
        return 0
    
    #If we have solved it earlier, then return the result from memory
    if dp[item_count][capacity] != -1:
        return dp[item_count][capacity]
 
    #Otherwise, we solve it for the new combination and save the results in the memory
    if weights[item_count-1] <= capacity:
        dp[item_count][capacity] = max(
            values[item_count-1] + knapsackM(weights, values, capacity-weights[item_count-1], item_count-1, dp),
            knapsackM(weights, values, capacity, item_count-1, dp)
            )
        return dp[item_count][capacity]

    dp[item_count][capacity] = knapsackM(weights, values, capacity, item_count-1, dp)
    return dp[item_count][capacity]
    
# --------------------------------------------------------------------------------------------------------------------------------------------------
# TABULATION (BOTTOM UP APPROACH) {ITERATIVE APPROACH}
# The bottom-up solution, also known as the tabulation technique, is an iterative method of solving dynamic programming problems. 
# Here, we will create a 2-D array of size (items + 1) ∗∗ (capacity + 1) 
# --------------------------------------------------------------------------------------------------------------------------------------------------

def knapsackT(weight,value,capacity,item_count):
    dp = [[0 for j in range(capacity + 1)] for i in range(item_count + 1)]
    # base condition
    for i in range(len(dp)):
        for j in range(len(dp[0])):
            if (i == 0 or j == 0):
                dp[i][j] = 0
            
            # check if the weight is less than the capacity
            if (weight[i-1]<=j):
                dp[i][j] = max(value[i-1]+dp[i-1][j-weight[i-1]] , dp[i-1][j]) #here value is array that's why [i-1] while dp is matrix with dp[0][0] = 0
            
            # check if the weight is more than capacity
            if (weight[i-1]>j):
                dp[i][j] = dp[i-1][j]
    
    return dp[-1][-1]
            

# ----------------------------------------------------------------------------------------------------------------------------------------------------
if __name__ == '__main__':
    result1 = knapsackN(weight, value, capacity, item_count)
    print("Maximum value:", result1)
    result2 = knapsackM(weight, value, capacity, item_count,dp)
    print("Maximum value:", result2)
    result3 = knapsackT(weight, value, capacity, item_count)
    print("Maximum value:", result3)
```
